```python
import matplotlib.pyplot as plt

# Initialize lists to hold epochs, training losses, and validation MSEs
epochs = []
train_losses = []
valid_mse = []

dir = "results of reviewer-old/XJTU results/0-0/Experiment1/logging.txt"

# Read the log file and parse the data
with open(dir, 'r') as log_file:
    for line in log_file:
        if '[Train]' in line:
            # Extract epoch and total loss from the training line
            parts = line.split(',')
            epoch = int(parts[0].split('epoch:')[1].strip())
            total_loss = float(parts[2].split('total loss:')[1].strip())
            
            epochs.append(epoch)
            train_losses.append(total_loss)
        elif '[Valid]' in line:
            # Extract MSE from the validation line
            mse = float(line.split('MSE:')[1].strip())
            valid_mse.append(mse)

# Plotting

# Figure 1: Training Loss
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, marker='o', label='Train Loss')
plt.yscale('log') 
plt.title('Training Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Figure 2: Validation MSE
plt.subplot(1, 2, 2)
plt.plot(epochs, valid_mse, marker='o', c='orange', label='Validation MSE')
plt.yscale('log')  
plt.title('Validation MSE over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.grid(True)
plt.legend()

# Display plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Initialize lists to hold epochs, training losses, and validation MSEs for both directories
epochs1, train_losses1, valid_mse1 = [], [], []
epochs2, train_losses2, valid_mse2 = [], [], []
epochs3, train_losses3, valid_mse3 = [], [], []

dir1 = "results of reviewer-old-200-epochs/XJTU results/0-0/Experiment1/logging.txt"
dir2 = "results of reviewer-new-with-wrong-loss-smaller/XJTU results/0-0/Experiment1/logging.txt"
dir3 = "results of reviewer-new-with-wrong-loss/XJTU results/0-0/Experiment1/logging.txt"

# Function to read log file and parse data
def parse_log_file(dir, epochs, train_losses, valid_mse):
    with open(dir, 'r') as log_file:
        for line in log_file:
            if '[Train]' in line:
                # Extract epoch and total loss from the training line
                parts = line.split(',')
                epoch = int(parts[0].split('epoch:')[1].strip())
                total_loss = float(parts[2].split('total loss:')[1].strip())
                
                epochs.append(epoch)
                train_losses.append(total_loss)
            elif '[Valid]' in line:
                # Extract MSE from the validation line
                mse = float(line.split('MSE:')[1].strip())
                valid_mse.append(mse)

# Parse both directories
parse_log_file(dir1, epochs1, train_losses1, valid_mse1)
parse_log_file(dir2, epochs2, train_losses2, valid_mse2)
parse_log_file(dir3, epochs3, train_losses3, valid_mse3)

# Plotting

# Figure 1: Training Loss Comparison
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs1, train_losses1, marker='o', label='Train Loss (Old)')
plt.plot(epochs2, train_losses2, marker='x', label='Train Loss (New)')
plt.plot(epochs3, train_losses3, marker='x', label='Train Loss (New-smaller)')
plt.yscale('log') 
plt.title('Training Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Figure 2: Validation MSE Comparison
plt.subplot(1, 2, 2)
plt.plot(epochs1, valid_mse1, marker='o', c='orange', label='Validation MSE (Old)')
plt.plot(epochs2, valid_mse2, marker='x', c='green', label='Validation MSE (New)')
plt.plot(epochs3, valid_mse3, marker='x', c='blue', label='Validation MSE (New-smaller)')
plt.yscale('log')  
plt.title('Validation MSE over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.grid(True)
plt.legend()

# Display plots
plt.tight_layout()
plt.savefig('comparison.png')
plt.show()
```
